```python
#Bai3
a = input("Nhap vao mot ky tu:")
if a.isupper():
    print("Chu thuong tuong ung la:",a.lower())
elif a.islower():
    print("Chu hoa tuong ung la:",a.upper())
else:
    print("Ban da khong nhap sai ky tu!",a)
```

```python
#Bai4
c = input("nhap vao mot ky tu:")
if c.isalpha():
    print(c,"la ki tu alphabet")
else:
    print(c,"khong phai la ki tu alphabet")
```

```python
#Bai5
a = input("nhap vao mot chu cai in hoa:")
if a == 'A':
    print(f"Day la truong hop dac biet: A la chu dau tien va khong co chu thuong lien truoc")
elif 'B' <= a <= 'Z':
    chuthuong = a.lower()
    ma_ascii = ord(chuthuong)
    ma_ascii_truoc = ma_ascii - 1
    ky_tu_truoc = chr(ma_ascii_truoc)
    print(ky_tu_truoc)
else:
    print("Dau vao '{a}' khong hop le!!!. Vui long nhap mot chu cai hoa tu A -> Z" )
```
