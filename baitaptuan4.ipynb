```python
#W4A6
import math
try:
    user_input = input("Nhap a b: ")
    a, b = map(int, user_input.split())
    if a > b:
        a, b = b, a
    tong = 0
    for i in range(a, b + 1):
        la_nguyen_to = True
        if i < 2:
            la_nguyen_to = False
        else:
            for k in range(2, math.isqrt(i) + 1):
                if i % k == 0:
                    la_nguyen_to = False
                    break 
        if la_nguyen_to:
            tong += i
    print(f"Tong cac so nguyen to tu {a} den {b} la: {tong}")
except ValueError:
    print("Lỗi: Hãy nhập đúng 2 số nguyên cách nhau bởi dấu cách!")

```

```python
#W4A8
def is_palindrome(n):
    return str(n) == str(n)[::-1]

def reverse_add_palindrome(n):
    steps = 0
    while not is_palindrome(n):
        n += int(str(n)[::-1])
        steps += 1
    return steps, n

n = int(input())
steps, palindrome = reverse_add_palindrome(n)
print(steps, palindrome)
```

```python
#W4A9
def distinct_digits(num): #kiểm tra các chữ số có khác nhau hay không.
    s = str(num)
    return len(s) == len(set(s))

n = int(input("Nhap vao so nguyen n:"))
ds = [] #tạo một danh sách để lưu các số chính phương hợp lệ.
x = 1 #bắt đầu kiểm tra từ 1
while x * x <= n: #lặp khi x mũ 2 <= n vì các số chính phương không vượt quá n
    sq = x * x #tính số chính phương của x
    if distinct_digits(sq): #gọi hàm để khiểm tra các chữ số của sq có khác nhau không
        ds.append(str(sq)) #nếu số hợp lệ, thêm sq vào danh sách vừa tạo + chuyển sang chuỗi
    x += 1 #tăng x lên để xét số chính phương tiếp theo

if ds: #kiểm tra nếu có ít nhất 1 số hợp lệ
    print(" ".join(ds)) #ghép các phần tử trong ds, in ra cùng một dòng, cách nhau 1 dấu cách
else:
    print("no number")
```

```python
#W4A10
n = int(input("Nhap vao so nguyen duong n:"))
max_do_dai = 0
so_tot_nhat = 0
for i in range(1, n+1):
    x = i
    do_dai = 1
    while x > 1:
        if x % 2 == 0:
            x = x // 2
        else:
            x = x * 3 + 1
        do_dai += 1
    
    if do_dai > max_do_dai:
        max_do_dai = do_dai
        so_tot_nhat = i

print(f"So co day collaz dai nhat la:",so_tot_nhat)
print(f"Day collaz cua so do gom:", max_do_dai ,"chu so.")
```

```python
#W4A13
def tong_uoc(num):
    tong = 0
    for i in range(1, num):
        if num % i == 0:
            tong += i
    return tong

a = int(input("Nhập số thứ nhất: "))
b = int(input("Nhập số thứ hai: "))

# Kiểm tra điều kiện số thân thiết
if tong_uoc(a) == b and tong_uoc(b) == a:
    print("true")
else:
    print("false")
    
```

```python
#W4A15
tong_con = int(input("Nhập tổng số con: "))
tong_chan = int(input("Nhập tổng số chân: "))
found = False

for gà in range(tong_con + 1):
    chó = tong_con - gà
    if (gà * 2 + chó * 4) == tong_chan:
        print(f"Gà: {gà}, Chó: {chó}")
        found = True
        break

if not found:
    print("invalid")
```
