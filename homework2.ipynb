```python
#bai1
a = float(input("Chieu dai hcn la:"))
b = float(input("Chieu rong hinh chu nhat la:"))
if a > b and a != 0 and b != 0:
    print(f"Dien tich hinh chu nhat la: {a * b}")
    print(f"Chu vi hinh chu nhat la: {(a + b) * 2}")
else:
    print("Khong phai la hinh chu nhat!!!")
```

```python
#bai2
n = float(input("Nhap vao chieu rong cua hinh tron:"))
if n != 0 and n > 0:
    print(f"Dien tich hinh tron la:{ n * n * 3.14 }")
    print(f"Chu vi hinh tron la: {2 * 3.14 * n}")
else:
    print("Ban da nhap sai gia tri!!!")
```
