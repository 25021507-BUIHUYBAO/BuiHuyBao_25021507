```python
#Bai4
h = int(input("Nhập độ cao h: "))
for i in range(1, h + 1):
    if i == 1:
        print(" " * (h - 1) + "*")
    elif i == h:
        print("*" * (2 * h - 1))
    else:
        khoang_trang_ngoai = " " * (h - i)
        khoang_trang_trong = " " * (2 * i - 3)
        print(khoang_trang_ngoai + "*" + khoang_trang_trong + "*")
```

```python
#Bai14
max_value = None
min_val = None
while True:
    num = int(input("Nhap so (nhap -1 de dung): "))
    if num == -1:
        break
    if max_value is None or num > max_value:
        max_value = num
    if min_val is None or num < min_val:
        min_val = num

if max_value is not None:
    print(f"So lon nhat: {max_value}, So nho nhat: {min_val}")
```

```python
#Bai16
n = abs(int(input("Nhập n: ")))
chan = 0
le = 0
if n == 0: chan = 1
while n > 0:
    chu_so = n % 10
    if chu_so % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10
print(f"Số chữ số chẵn: {chan}, Số chữ số lẻ: {le}")
```

```python
#Bai18
n = int(input("Nhập n: "))
is_3k = False
if n > 0:
    temp = n
    while temp % 3 == 0:
        temp //= 3
    if temp == 1:
        is_3k = True

if is_3k:
    print(f"{n} có dạng 3^k")
else:
    print(f"{n} không có dạng 3^k")
```

```python
#Bai19
A = int(input("Nhập A: "))
f1, f2 = 1, 1
fib_max = 1
while f2 <= A:
    fib_max = f2
    f1, f2 = f2, f1 + f2
print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {fib_max}")
```

```python
#Bai20
n = abs(int(input("Nhập n: ")))
count = 0
while n > 0:
    chu_so = n % 10
    if chu_so in (2, 3, 5, 7):
        count += 1
    n //= 10
print(f"Số lượng chữ số là số nguyên tố: {count}")
```
