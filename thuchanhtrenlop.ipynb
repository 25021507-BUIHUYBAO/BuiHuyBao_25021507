```python
#tamgiac
import math
a , b , c = map(float,input().split())
if (a > 0 and b > 0 and c > 0) and (a + b > c) and (a + c >b) and (b + c > a):
    print("do la tam giac thuong")
elif (a == b and b != c) or (a == c and c != b) or (b == c and c!= a):
    print("do la tam giac can")
elif a**2 + b**2 == c**2 or b**2 + c**2 == a**2 or a**2 + c**2 == b**2:
    print("do la tam giac vuong")
elif a == b == c:
    print("do la tam giac deu")
else:
    print("khong phai la tam giac")    
```

```python
lst = input("Nhap mot day so").split()
n = []
for i in lst:
    so = int(i)
    n.append(so)
tong = 0
for x in n:
    tong = tong + x
soluong = len(n)
if soluong > 0:
    print(f"Trung binh cua day so la: {tong / soluong}")
else:
    print("Nhap loi!!!")

```
